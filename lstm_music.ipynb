```python
import numpy as np
import sys
#读取文件，并全部变成小写
filename = "gaobai.txt"
raw_text = open(filename).read()
raw_text = raw_text.strip()
raw_text = raw_text.lower()
print(raw_text)
```

```python
chars = sorted(list(set(raw_text)))   #文本出现的所有字符 27个
char_to_int = dict((c, i) for i, c in enumerate(chars))   #字符串对应相应的id
int_to_char = dict((i, c) for i, c in enumerate(chars))   #id对应相应的字符串

n_chars = len(raw_text)
n_vocab = len(chars)
print(chars)
n_vocab
```

```python
seq_length = 1  #设置步长
dataX = []
dataY = []
for i in range(0, n_chars - seq_length, 1):
    seq_in = raw_text[i:i + seq_length]
    seq_out = raw_text[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
n_patterns = len(dataX)
print ("Total Patterns: ", n_patterns)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, Dropout
from keras.utils import np_utils

X = np.reshape(dataX, (n_patterns, seq_length, 1))  #3D张量
X = X/float(len(raw_text))
y = np_utils.to_categorical(dataY)  #one hot encode

# create and fit the model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=2)

# summarize performance of the model
scores = model.evaluate(X, y, verbose=0)
print("Model Accuracy: %.2f%%" % (scores[1]*100))
```

```python
# generate characters
for pattern in dataX:
    x = np.reshape(pattern, (1, len(pattern), 1))
    x = x / float(n_vocab)
    prediction = model.predict(x, verbose=0)
    index = np.argmax(prediction)  #返回最大值的索引
    result = int_to_char[index]
    seq_in = [int_to_char[value] for value in pattern]
    print (seq_in, "->", result)

print ("\nDone.")
```
